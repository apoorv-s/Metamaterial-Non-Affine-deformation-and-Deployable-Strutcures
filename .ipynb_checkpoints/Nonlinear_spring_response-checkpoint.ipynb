```python
x1 = 0
y1 = 0
x2 = 1
x3 = 2
y3 = 0

k1 = 1
k2 = 1
l0_1 = 1
l0_2 = 1

del_vec = []
R_vec = []

for j in range(0, 300):
    y2 = j/200
    del_vec.append(y2)
    delta_l1 = Length(x1, y1, 0, x2, y2, 0) - l0_1
    delta_l2 = Length(x3, y3, 0, x2, y2, 0) - l0_2
    F1 = k1*delta_l1
    F2 = k2*delta_l2
    theta1 = np.arctan(y2/l0_1)
    theta2 = np.arctan(y2/l0_2)
    
    R = (F1*np.sin(theta1)) + (F2*np.sin(theta2))
    
    R_vec.append(R)
```

```python
plt.plot(del_vec, R_vec)
plt.xlabel('Reaction (normalized by k)') 
plt.ylabel('Delta (normalized by $l_0$)')
plt.savefig('spring_response.png', dpi=300)
plt.show()
```
